```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
```

```python
csv_file = "Resources/dogNames2.csv"
dog_data_df = pd.read_csv(csv_file)
dog_data_df.head()
```

```python
csv_file = "Resources/allBabyNamesUSA_2000s.csv"
baby_df = pd.read_csv(csv_file)
baby_df.head()
```

```python
baby_df = baby_df.rename(columns={"Number": "Count_of_Names"})
baby_df.head()
```

```python
baby_df.drop(['Gender'], axis=1, inplace=True)
```

```python
baby_df.drop(['Rank'], axis=1, inplace=True)
```

```python
baby_df.head()
```

```python
baby_df = baby_df.groupby(['Name']).sum()
baby_df.head()
```

```python
new_df = dog_data_df.rename(columns={"Row_Labels": "Name", "Count_AnimalName": "Count_of_Names"})
new_df.head(10)
```

```python
clean_df = new_df[~new_df['Name'].isin(['1','2','40804','90201','90203','102201','3010271'])]
clean_df.head()
```

```python
clean_df['Name'] = clean_df['Name'].str.capitalize()
```

```python
clean_df.head()
```

```python
merge_df = pd.merge(baby_df, clean_df, how='inner', on='Name')
```

```python
merge_df.set_index('Name', inplace=True)
```

```python
merge_df.head()
```

```python
merge_df = merge_df.rename(columns={"Count_of_Names_x": "Count of Baby Names", "Count_of_Names_y": "Count of Dog Names"})
```

```python
merge_df.head()
```

```python
merge_df.to_csv('Name_comparison.csv')
```
